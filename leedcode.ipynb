```python
'''  1) Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M  '''
def roman_to_int(s):
    roman = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }
    total = 0

    for i in range(len(s)):
        if i + 1 < len(s) and  roman[s[i]] < roman[s[i+1]]:
            total -= roman[s[i]]
        else:
            total += roman[s[i]]
    return total

while True:
    s = input("Roma numeral: ").upper()
    print("Result: ", roman_to_int(s))
```

```python
'''2) Given an integer array nums, return the length of the longest strictly increasing subsequence.'''
def length_of_lst(nums):
    n = len(nums)
    dp = [1] * n  

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
lst=length_of_lst([10, 9, 2, 5, 3, 7, 101, 18])
print("Result: ", lst)
```
