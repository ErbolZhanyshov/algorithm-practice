```python
'''  1) Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M  '''
def roman_to_int(s):
    roman = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }
    total = 0

    for i in range(len(s)):
        if i + 1 < len(s) and  roman[s[i]] < roman[s[i+1]]:
            total -= roman[s[i]]
        else:
            total += roman[s[i]]
    return total

while True:
    s = input("Roma numeral: ").upper()
    print("Result: ", roman_to_int(s))
```

```python
'''2) Given an integer array nums, return the length of the longest strictly increasing subsequence.'''
def length_of_lst(nums):
    n = len(nums)
    dp = [1] * n  

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
lst=length_of_lst([10, 9, 2, 5, 3, 7, 101, 18])
print("Result: ", lst)
```

```python
'''3) Using prefix sum to efficiently answer multiple range sum queries.'''
class NumArray:
    def __init__(self, nums):
        self.prefix = [0]

        running_sum = 0
        for num in nums:
            running_sum += num
            self.prefix.append(running_sum)

    def sumRange(self, left, right):
        # sum from left to right = prefix[right+1] - prefix[left]
        return self.prefix[right + 1] - self.prefix[left]


nums = [1, 2, 3, 4, 5]
obj = NumArray(nums)

print(obj.sumRange(1, 3))  
print(obj.sumRange(0, 4))  
```

```python
'''Intuition
We choose the first two numbers and recursively check whether the rest of the string follows the additive sequence rule.'''

class Solution:
    def isAdditiveNumber(self, num:str)->bool:
        def check(a,b,rest):
            if(len(a)>1 and a[0]=='0') or (len(b)>1 and b[0]=='0'):
                return False
            
            x = int(a)
            y = int(b)

            while rest:
                s = str(x+y)
                if not rest.startswith(s):
                    return  False
                rest = rest[len(s):]
                x = y
                y = int(s)
            return True
        n = len(num)
        for i in range(1, n):
            for j in range(i+1, n):
                first = num [:i]
                second = num[i:j]
                remaining = num[j:]

                if check(first, second, remaining):
                    return True
        return False
while True:
        z = input("Sayi gir: ")
        addi = Solution() 

        if addi.isAdditiveNumber(z):
            print("Additive sayi")
        else:
            print("Degil")
```

```python
'''Given a string s containing only '(', ')', '{', '}', '[', and ']', determine if the input string is valid.'''
def isValid(s):
        stack = []

        pairs = {
            ')': '(',
            '}': '{',
            ']': '['
        }

        for i in s:
            if i in "({[":
                stack.append(i)
            else:
                if not stack:
                    return False

                if stack[-1] == pairs[i]:
                    stack.pop()
                else:
                    return False

        return len(stack) == 0

print(isValid("()"))        
print(isValid("()[]{}"))    
print(isValid("(]"))        
print(isValid("([)]"))      
print(isValid("{[]}"))      
print(isValid("((("))       
```
